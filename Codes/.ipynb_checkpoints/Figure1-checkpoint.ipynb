```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = {
        "dark_purple": "#5F2E88",
        "purple": "#7E59A2",
        "light_purple": "#A17DB8",
        "pale_purple": "#dfd6e5",
        "dark_orange": "#F38227",
        "orange": "#E39943",
        "light_orange": "#EEBA7F",
        "pale_orange": "#f2d4b6",
        "dark_blue": "#3F60AC",
        "blue": "#7292C7",
        "light_blue": "#A5B3CC",
        "pale_blue": "#dae4f1",
        "dark_red": "#9C372F",
        "red": "#C76A6A",
        "light_red": "#E39C9D",
        "pale_red": "#edcccc",
        "dark_green": "#395A34",
        "green": "#688A2F",
        "light_green": "#B3CD86",
        "pale_green": "#d8e2c3",
        "dark_brown": "#764f2a",
        "brown": "#c2996c",
        "light_brown": "#e1bb96",
        "pale_brown": "#efccaf",
        "black": "#444147",
        "light_grey": "#EFEFEF",
        "light_gray": "#EFEFEF",
        "grey": "#6D6F72",
        "gray": "#6D6F72",
}
```

```python
# Generate a truncated exponential for 'Radical' by using an exponential distribution and discarding values above 1
#radical = np.random.exponential(scale=0.2, size=10000)
#radical = radical[radical <= 1]
radical=np.abs(np.random.normal(loc=0, scale=0.1, size=1000))


# Adjusting the 'Moderate' histogram to be slightly wider
moderate = np.random.triangular(left=0.45, mode=0.5, right=0.55, size=1000)  # Wider peak around the middle
#conservative = np.random.beta(a=5., b=1., size=1000) # left-skewed distribution
conservative=1 - np.abs(np.random.normal(loc=0, scale=0.1, size=1000))
random = np.random.uniform(low=0, high=1, size=1000)  # uniform distribution
# Generate a 'Bimodal' distribution with half-normal peaks at both ends (0 and 1)
bimodal_left = np.abs(np.random.normal(loc=0, scale=0.1, size=500))  # Half-normal distribution for the left peak
bimodal_right = 1 - np.abs(np.random.normal(loc=0, scale=0.1, size=500))  # Inverted half-normal distribution for the right peak
bimodal = np.concatenate([bimodal_left, bimodal_right])
titles = ['Radical', 'Moderate', 'Conservative', 'Random', 'Bimodal']
data = [radical, moderate, conservative, random, bimodal]

# Adjusting the sample data ranges and removing y-ticks and grids

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Remove the last (empty) subplot
fig.delaxes(axs[1,2])

# Plot histograms with x-axis limited between 0 and 1, remove y-ticks and grids
for ax, hist_data, title in zip(axs.flat, data, titles):
    ax.hist(hist_data, bins=30, color=colors['gray'], edgecolor='black', density=True, range=(0, 1))
    ax.set_title(title,fontsize=25)
    ax.set_xlabel('Fitness quantile',fontsize=20)
    ax.set_ylabel('Probability density',fontsize=20)
    ax.set_xlim(0, 1)
    ax.set_yticks([])
    ax.tick_params(axis='both', which='major', labelsize=15)
    ax.grid(False)

# Adjust the layout
plt.tight_layout()
plt.show()
```
